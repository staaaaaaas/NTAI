```python
def get_noise_data(X, M = 0.2):
    X_noisy = np.zeros((X.shape[0], X.shape[1]))

    for i in range(X.shape[1]):
        if np.random.rand() < M:
            X_noisy[0, i] = - X[0, i]
        else:
            X_noisy[0, i] = X[0, i]            
    return X_noisy
```

```python
zero_noisy = get_noise_data(zero, M = 0.1)
plt.figure(figsize=(4, 6))
plt.title('Zero noisy param 0.2')
plt.imshow(zero_noisy.reshape(-1).reshape((12,10)))
plt.show()
```

```python
six_noisy = get_noise_data(six, M = 0.15)
plt.figure(figsize=(4, 6))
plt.title('Six noisy param 0.3')
plt.imshow(six_noisy.reshape(-1).reshape((12,10)))
plt.show()
```

# Сеть Хэмминга

```python
zero = zero.reshape(-1)
one = one.reshape(-1)
two = two.reshape(-1)
three = three.reshape(-1)
four = four.reshape(-1)
six = six.reshape(-1)
nine = nine.reshape(-1)

zero_noisy = zero_noisy.reshape(-1)
six_noisy = six_noisy.reshape(-1)
```

```python
patterns = np.array([zero, one, two, three, four, six, nine])
Q = len(patterns)
eps = 1 / (Q - 3)
R = 12 * 10
```

```python
IW = np.array([zero.T, one.T, two.T, three.T, four.T, six.T, nine.T])
print(f'IW shape: {IW.shape}')
b = R * np.ones((Q, 1))
print(f'b shape: {b.shape}')
```

```python
a1 = np.zeros((Q, Q))

for i in range(Q):
    a1[i] = IW[i] @ patterns[i] + b[i]

print(f'a1 shape: {a1.shape}')
```

```python
LW = np.eye(Q)
LW[LW == 0.0] = -eps
print(f'LW shape: {LW.shape}')
```

```python
from neurolab.net import newhop

model = newhop(a1, max_init=600)

model.layers[0].np['w'][:] = LW
model.layers[0].np['b'][:] = 0
```

## Тест для 1, без зашумления

```python
one_a1 = IW @ one + b
one_a1
```

```python
one_res = model.sim(one_a1)
best_ind = np.argmax(one_res[0])

plt.figure(figsize=(4, 6))
plt.title('Answer')
plt.imshow(IW[best_ind].reshape(-1).reshape((12,10)))
plt.show()
```
